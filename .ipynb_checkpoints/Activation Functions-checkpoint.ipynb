```python
import math
```

```python
def sigmoid(x):
    return 1/(1+math.exp(-x))
```

```python
sigmoid(100)
```

```python
sigmoid(1)
```

```python
sigmoid(-56)
```

```python
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x)+math.exp(-x))
```

```python
tanh(-56)
```

```python
tanh(2)
```

```python
def reLu(x):
    return max(0,x)
```

```python
reLu(8)
```

```python
reLu(-3)
```

```python
def LeakyReLu(x):
    return max(0.1*x,x)
```

```python
LeakyReLu(5)
```

```python
LeakyReLu(-3)
```
